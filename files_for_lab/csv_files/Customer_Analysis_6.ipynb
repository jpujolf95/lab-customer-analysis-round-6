```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv('marketing_customer_analysis.csv') 
data.head()
```

```python
data = data.drop(['Customer','Effective To Date'], axis=1)
```

1. X-Y Split

```python
y = data['Total Claim Amount']
X = data.drop(['Total Claim Amount'], axis=1)
X.head()
```

2. Encoding the Categoricals

```python
X_num = X.select_dtypes(include = np.number) #filter numerical and categorical data in each data frame
X_cat = X.select_dtypes(include = 'object')
```

```python
X = pd.get_dummies(X) # You can put X, not separating variables is needed.
X.head()
```

3. Train-test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

4. Predictions

```python
predictions  = model.predict(X_test) #predict on test sample
predictions
```

```python
y_test
```

5. Model, is it any good?

```python
r2 = r2_score(y_test, predictions)
RMSE = mean_squared_error(y_test, predictions, squared=False)
MSE = mean_squared_error(y_test, predictions)
print("r2 = ", r2) 
print("RMSE = ", RMSE) 
print("MSE = ", MSE)
```
